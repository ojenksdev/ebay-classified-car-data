The first change I made to the dataframe had to do with the column names. Some of them were more difficult to understand and needed to be renamed. Also, the original columns used camelcase instead of snake case, so I changed that as well for readability.

```python
autos.describe(include="all")
```

Looking at the data thus far, it looks like a column we can probably drop is the **nr_of_pictures** as the value is mostly 0.0 across the rows. Another column I'm not too sure about is the **ab_test** column. I don't know how useful that column will be to our research. **unrepaired_damage** might be able to be dropped as well. 

We'll need to convert the **price** and **odometer** columns into floats.

```python
autos["price"].head()
```

```python
autos["price"] = (autos["price"]
                 .str.replace("$", "")
                 .str.replace(",", "")
                 .astype(float))
```

```python
autos["price"].head()
```

**Now we can go about converting the odometer column:**

```python
autos["odometer"].describe()
```

```python
autos["odometer"].head(10)
```

**_We can see there is a str 'km' and a comma that need to be removed before we can convert:_**

```python
autos["odometer"] = (autos["odometer"]
                    .str.replace(",", "")
                    .str.replace("km", ""))
autos["odometer"].head()
```

**_Now we can convert the strings over to type float:_**

```python
autos["odometer"] = autos["odometer"].astype(float)
```

```python
autos["odometer"].head()
```

**_There we go! Converted over to floats. Next thing we need to do is rename the column since we removed the km from data:_**

```python
autos.rename({"odometer":"odometer_km"}, axis=1, inplace=True)
```

**_Odometer is taken care of now! Lets move on._**

```python
autos["odometer_km"].describe()
```

I think the odometer readings are pretty good, especially for used cars. I don't see any range or amount of miles that is out of the ordinary, so I will move on to the price column.

```python
autos["price"].value_counts().sort_index(ascending=True)
```

**Ok, while it is 'possible' to be selling a used car for over 100k, I'm not sure if we'd find those in ebay classifieds. Time to look a bit deeper into the data to examine what kind of cars these are.**

**The same goes for the cars that are 'free'**

```python
free_cars = autos[autos["price"] == 0.0]
free_cars.head()
```

As I thought, these "free" vehicles aren't needed in our data because we have no way of truly knowing why they are labeled free. My guess is the person who added them to the classifieds either forgot to add the price (which is rare, but I guess possible), or they are trying to lure in potential buyers in an odd fashion. Either way, this data will be removed. 

```python
autos = autos[autos["price"] > 0.0]
autos["price"].value_counts().sort_index(ascending=True)
```

**Ok, we've removed all of the free vehicles from the data. Now, lets take out the car listings with a price over 350k as the next price after that is just shy of a million dollars:**

```python
pricey_cars = autos[autos["price"] > 350001]
pricey_cars.head(5)
```

**Now this is something I wasn't expecting to see. It does appear some people are selling cars that are luxury and collectors items for extremely high prices. However, because the amount of cars is so low, we'll still remove them**

```python
autos = autos[autos["price"] < 350001]
```

```python
autos["price"].value_counts().sort_index(ascending=True)
```

It looks like most of the cars listed were registered around the year 2000. Quite a few vehicles were listed as being registered in 2016, but I wonder if that is simply based around the date last_seen online. I noticed that the registration_month is 0 for quite a few entries.

```python
autos["brand"].value_counts().sort_index(ascending=False)
```

What I want to take a look at is the 5 most popular vehicles to see if there is any pattern linking their popularity. **Volkswagon** is clearly the most popular used car for sale. Next up is **bmw** and then **opel**. We close out the list with **mercedez_benz** and **audi**. 

```python
most_pop_on_sale = {}
popular_cars = ["volkswagen", "bmw", "opel", "mercedes_benz", "audi"]

for c in popular_cars:
    car = autos[autos["brand"] == c]
    mean = car["price"].mean()
    most_pop_on_sale[c] = mean

print(most_pop_on_sale)
    
```

After running the loop on the selected vehicles, it appears that the most popular cars have an average price under 10,000 dollars. This might be the reason for the popularity. 

Another thing I wonder is if sellers are going for a lower price than what the vehicles are actually worth. What makes the Audi vehicles worth more than the Mercedes-Benz? And are the Volkswagen vehicles priced to sell due to other reasons? We would need more data to find out.

```python
pop_car_mileage = {}

for c in popular_cars:
    car = autos[autos["brand"] == c]
    mean = car["odometer_km"].mean()
    pop_car_mileage[c] = mean
 
pop_car_mileage
    
    
```

Now lets convert these two dictionaries into objects:

```python
pop_prices_series = pd.Series(most_pop_on_sale)
print(pop_prices_series)
```

```python
pop_cars = pd.DataFrame(pop_prices_series, columns=["mean_prices"])
pop_cars
```

```python
pop_mileage_series = pd.Series(pop_car_mileage)
pop_cars["mean_mileage"] = pop_mileage_series
pop_cars
```

### Findings:
Looking at this dataframe for the most popular cars, its clear the mileage isn't necessarily playing a factor in the cost of the vehicle. **bmw** appears to have an average mileage of 132k _(highest in the bunch)_, but comes in at the second highest average cost _($8,332)_. 

Looking at this df, the **volkswagen** and **opel** seem to be the best value, price and mileage-wise. However, it would be nice to have seen some data regarding the # of accidents, recalls, avg cost of repairs, for each brand to get a true sense of the value for a buyer.
